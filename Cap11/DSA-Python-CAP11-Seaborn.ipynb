```python
#Criando gr√°ficos bivariados e univariados usando o joinplot
import seaborn as sns
sns.jointplot(data=dados, x="total_bill", y="tip", kind='reg') #gr√°fico do tipo reg com uma linha de regress√£o
```

## Usando Matplotlib, Seaborn, Numpy e Pandas na Cria√ß√£o de gr√°fico estat√≠stico

```python
#Valores rand√¥micos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

#Vari√°veis
flag_fumante = np.random.rand(n) <pct_smokers
idade = np.random.normal(40,10,n)
altura = np.random.normal(170, 10, n)

peso = np.random.normal(70,10,n)

#DataFrame
dados = pd.DataFrame({'altura': altura, 'peso':peso, 'flag_fumante':flag_fumante})

#Cria os dados para a vari√°vel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True:'Fumante',False:'N√£o Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
#Style
sns.set(style = "ticks")

#lmplot
sns.lmplot (x = 'altura',
            y = 'peso',
            data = dados,
            hue = 'flag_fumante',
            palette =['tab:blue', 'tab:orange'],
            height =7
           )

#labels e t√≠tulo
plt.xlabel('Altura(cm)')
plt.ylabel('Peso (kg)')
plt.title('Rela√ß√£o entre Altura e Peso de Fumante e N√£o Fumante')

#Remove bordas
sns.despine()

#show
plt.show()
```
